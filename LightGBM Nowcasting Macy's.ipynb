# Light GBM

```python
from lightgbm import LGBMRegressor
```

```python
def LGBMregression4(X2,Y2):
    from sklearn.model_selection import train_test_split

    X_train, X_test, Y_train, Y_test = train_test_split(X2, Y2, test_size=0.20, random_state=42)
    
    model = LGBMRegressor(boosting_type='gbdt', objective='regression',max_bin=2000, num_leaves=3000,learning_rate=0.07, n_estimators=500,num_iteration=1000, max_depth=5000,metric='rmse', bagging_fraction=0.9, feature_fraction=0.8, reg_lambda=0.7)

    model.fit(X_train,Y_train)

    predictions = model.predict(X_test)

    %matplotlib inline
    import matplotlib.pyplot as plt
    plt.style.use('seaborn-whitegrid')
    import numpy as np
    plt.rcParams["figure.dpi"] = 300
    fig = plt.figure()
    ax = plt.axes()

    x = np.linspace(0, len(predictions))
    ax.plot(list(predictions),linewidth=1,color="blue")
    ax.plot(list(Y_test),linewidth=1,color="orange")

    from sklearn.metrics import mean_squared_error
    from math import sqrt
    rms = sqrt(mean_squared_error(Y_test, predictions))
    print (rms)
    
```

```python
LGBMregression4(X2,Y2)
```
